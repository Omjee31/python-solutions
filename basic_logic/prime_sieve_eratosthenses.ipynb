Generate prime numbers using Sieve of Eratosthenes.

```python
def prime_check(num):
    prime = []

 
    for n in range(1,num):
        if num<2:
            continue
        is_prime = True
        for i in range(2,int(n**0.5)+1):
            if n%i==0:
                is_prime = False
                break
        if is_prime:
            prime.append(n)
    return f"Prime By sieve of Eratosthenes= {prime}"
    


print(prime_check(35))
```
