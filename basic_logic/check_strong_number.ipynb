```python
import math

def check_strong_number():
    # Loop through numbers from 1 to 999
    for i in range(1, 1000):
        n = i  # Store current number
        
        # Convert number to string to access each digit
        # Convert each digit back to int
        # Take factorial of each digit
        # Sum all factorial values
        all_number_ch = sum(math.factorial(int(d)) for d in str(n))
        
        # Check if original number equals sum of factorial of its digits
        if n == all_number_ch:
            # If condition is true, print the number
            print(n, end=' ')

# Call the function
check_strong_number()
```
