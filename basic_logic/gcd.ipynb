12 Find GCD & LCM using recursion.


```python
def gcd(num1, num2):
    l1 = [i for i in range(1, num1 + 1) if num1 % i == 0]
    l2 = [i for i in range(1, num2 + 1) if num2 % i == 0]

    common = [i for i in l1 if i in l2]

    return max(common)

```

```python
print(gcd(15,16))
```

Take two numbers, a and b

Divide a by b and note the remainder r.

Replace a with b and b with r.

Repeat the process until the remainder becomes 0.

The last non-zero divisor is the GCD.

```python
def gcd_eulidiean(a, b):
    a = abs(a)
    b = abs(b)
    
    while b != 0:
        a, b = b, a % b
    return a
```

```python
print(gcd_eulidiean(34,56))
```

```python
def gcd_recursive(a,b):
    if b == 0:
        return a
    return gcd_recursive(b,a%b)
    
```

```python
print(gcd_recursive(34,56))
```

```python
def lcm(num1,num2):
    return (num1*num2)//gcd_eulidiean(num1,num2)


print(lcm(12, 18))   # 36
print(lcm(5, 10))    # 10
print(lcm(7, 9))    # 63
```
