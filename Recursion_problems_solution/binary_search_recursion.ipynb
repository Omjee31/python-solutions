 Binary search using recursion.

```python
def binary(arr, left, right, target):
    if left > right:
        return -1
        
    mid = (left + right) // 2
    
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary(arr, mid + 1, right, target)  # RETURN added
    else:
        return binary(arr, left, mid - 1, target)   # RETURN added


def main():
    l1 = list(map(int, input("Enter values with spacing: ").split()))
    target = int(input("Enter value you want to search: "))
    
    l1.sort()
    
    result = binary(l1, 0, len(l1) - 1, target)
    
    if result != -1:
        print(f"Element found at index {result}")
    else:
        print("Element not found")


if __name__ == "__main__":
    main()
```
