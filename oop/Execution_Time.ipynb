```python
import time   # used to measure execution time

class Operation:
    
    # Constructor â†’ runs when object is created
    def __init__(self, filename):
        self.filename = filename   # file name to read
        self.freq = {}             # dictionary to store word frequency
    
    # Method to read file and count word frequency
    def file_read(self):
        # Open file in read mode
        with open(self.filename, 'r') as f:
            
            # Read whole file and convert to lowercase
            self.words = f.read().lower()
            
            # Split text into list of words
            self.words = self.words.split()
            
            # Count frequency of each word
            for word in self.words:
                self.freq[word] = self.freq.get(word, 0) + 1
    
    # Method to display frequency result
    def show_result(self):
        for wordd, frequency in self.freq.items():
            #print(wordd, '--', frequency)

    # Method to measure execution time of other methods
    def show_execution_time(self):
        
        # Measure time for file_read()
        start = time.time()
        self.file_read()
        end = time.time()
        print(f'Time Execution Of read File is {end - start}')
        
        # Measure time for show_result()
        start = time.time()
        self.show_result()
        end = time.time()
        
        return f'Time Execution Of show result is {end - start}'


# Create object and pass filename
ope_1 = Operation('demo.txt')

# Call method
print(ope_1.show_execution_time())
```
