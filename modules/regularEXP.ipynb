```python
import re
string = "i am omjee singh from bihar mob is 9508664633 dob is 31"
print(re.findall('om',string))
```

```python
print(re.findall('\d',string))
print(re.findall('\d+',string),"With '+' shows all numbers in group")
```

| Method                      | When to Use                                   |
| --------------------------- | --------------------------------------------- |
| `re.findall(pattern, text)` | One-time quick use                            |
| `re.compile()`              | Repeated use, loops, large data, cleaner code |


```python
p = re.compile('[a-f]')
print(p.findall(string))
print(re.findall('[a-f]',string))
```

```python
p = re.compile('[a-f]')
for i in range (10):
    print(p.findall(string))
```

```python
p = re.compile('\d')
p1 = re.compile('\d+')
print(p.findall(string))
print(p1.findall(string))
```

 * \w  -- single charector in output
 * \w+ -- group charector(word) in output
 * \W  -- except word(non - word) special  charecter in output

```python
p,p1,p2 = '\w','\w+','\W'
string = "i am omjee singh # from @  bihar mob is 9508664633 dob is 31"

print(re.findall(p,string))
print(re.findall(p1,string))
print(re.findall(p2,string))
```

* re.split(pattern,string,maxsplit,flag=IGNORECASE)

```python
pattern = '\W+'
pattern2 = '\W'
print(re.split(pattern,string))
print(re.split(pattern2,string))
```
